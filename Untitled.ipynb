```python
import matplotlib.pyplot as plt
import numpy as np

x =[int(i) for i in range(1,5)]
p = [0.671,.620,0.567,0.512]

lx = []
px = []

n = 3

x0 = np.array([1.5,2.5,3.5])
x_1 = np.array(x)
f = np.array(p)

for k in x0:
    y_res = 0
    
    for i in range(0,n+1):
        
        z = 1.0
        
        for j in range(0,n+1):
            if j != i:
                z *= (k -x[j]) / (x[i] - x[j])
        y_res += z*f[i]
    print('El polinomio evualdo en f({0:}) = {1:} '.format(k,y_res))
    lx.append(k)
    px.append(y_res)
print(lx)
print(px)

x += lx
p += px
plt.plot(x,p,'bo')
```
