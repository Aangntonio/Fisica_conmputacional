f(x_0 + h) \approx f(x_0) + f^{\prime}(x_0)h +f^{\prime\prime}(x_0)\dfrac{h^2}{2}

$$
f^{\prime}(x_0) = \dfrac{f(x_0+h) - f(x_0) }{h} + \mathcal{O}
$$

```python
def derivada(h):
    derivada = (np.sin(2*np.pi*(0.45 + h)) - np.sin(2*np.pi * 0.45)) / h
    return derivada

    
```

```python
def derivada(h):
    f_x = (np.sin(2*np.pi*(0.45 + h)) - np.sin(2*np.pi * 0.45)) / h
    return f_x
    #pass

exacto = -5.97566
h = 1e-1
print('h \t Valor \t\t Error')
print('-'*50)
lista_h = []
lista_error = []

while h >1e-16:
    f_evaluado = derivada(h)
    error_relatov =  abs(error(exacto,f_evaluado))/100
    print('{0:1.0e} \t {1:1.10f} \t {2:1.5e}'.format(h,f_evaluado,error_relatov))
    lista_h.append(h)
    lista_error.append(error_relatov)
    h = h/10
print(h)

```

```python

plt.plot(lista_h,lista_error,'r+')
plt.plot(lista_h,lista_error,'-',color = '#4F02DD')
plt.xscale('log')
plt.yscale('log')
plt.xlabel('h')
plt.ylabel('Error')
plt.show()
```

```python
lista_reves = [lista_h[i] for i in range(len(lista_h) - 1, -1, -1)]
lista_reves_error = [lista_error[i] for i in range(len(lista_h) - 1, -1, -1)]
plt.plot(lista_reves,lista_reves_error,'r+')
plt.plot(lista_reves,lista_reves_error,'-',color = '#4F02DD')
plt.xlabel('h')
plt.ylabel('error')
plt.xscale('log')
```

```python
def barkara(a,b,c):
    '''
    Esta funciÃ³n puede resolver una ecuacion cuadratica con :
    a: 
    b:
    c:
    '''
    if b**2- 4 * a * c <0:
        raise ValueError('La soluciÂ´pon no esta definida en R')
    if a == 0:
        raise ValueError('a debe ser distindo de 0')
        
    dis = np.sqrt(b**2 - 4*a*c)
    x1 = (-b+  dis)/ (2*a)
    x2 = (-b-  dis)/(2*a)
    
    f = lambda x: a*x**2 + b*x + c 
    x = np.linspace(-10,10,101)
    
    plt.plot(x,f(x))
    plt.axhline(0, color='black', linewidth=0.5)
    plt.axvline(0, color='black', linewidth=0.5)
    plt.scatter(x1, 0, color='red', marker='o')
    cadena = 'las soluciones a tu ecuacion son' + str(x1) +' y ' + str(x2)
    plt.scatter(x2, 0, color='red', marker='o',label = cadena)
    plt.grid()
    plt.legend()
    plt.show()
    
    
    return cadena 
```
