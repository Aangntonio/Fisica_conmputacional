```python

import numpy as np
import matplotlib.pyplot as plt
```

Para la expansi칩n de $ln(x)$ se puede ver por series de talyor como
$$
ln(x+1) = \displaystyle{\sum_{i=1}^n \dfrac{(-1)^{n+1} x^n}{n}  }
$$
en particular si x= 1 para $ln(2)$
$$
ln(2) = \displaystyle{\sum_{i=1}^n  \dfrac{(-1)^{n+1}}{n}}
$$

Pero esa expansi칩n se puede ver como:
$$
ln 2 = 1 - \dfrac{1}{2} + \dfrac{1}{3} - \dfrac{1}{4} +...
$$
Se puede separar en su parte negativa y positiva:
$$
\begin{align}
ln 2 &= \left(1  + \dfrac{1}{3} + \dfrac{1}{5} + ...\right) - \left( \dfrac{1}{2} + \dfrac{1}{4} + ...\right)\\
ln 2 &= \displaystyle{\sum_{i=1}^n  \dfrac{1}{2n-1}} - \displaystyle{\sum_{i=1}^n  \dfrac{1}{2n}}
\end{align}
$$

```python
def error (arg1,arg2):
    '''
    Esta funci칩n genera el valor del error abosulto de 2 valores
    '''
    err = abs( (arg1 - arg2/ arg1)) * 100
    return err
```

```python

def ln_2(n):
    suma_pos = 0 #Guarda e침 valor de la suma de los numeros postivos
    suma_neg = 0 # Guarda el valor de la suma de los numeros negativos
    for i in range(1,n+1):
        suma_pos = suma_pos + 1/(2*i -1)
        suma_neg = suma_neg + 1/(2*i)
    return suma_pos - suma_neg
    
```

```python
t = 2
print('iteraciones \t Valor real \t\t Aproximacion \t\t Error')
print('-'*80)
lista_error = []
for l in range (0,150):
    valor1 = ln_2(l)
    error1 = abs(np.log(2) - valor1)/np.log(2)
    lista_error.append(error1)
    
    print('{0:} \t\t {1:1.10f} \t\t {2:1.10f} \t\t {3:1.5e} '.format(l,np.log(2),valor1,error1))

   # print(l,valor1,error1,valor2,error2)
```

```python
plt.plot(lista_error,'r+')
plt.plot(lista_error,'b-')
plt.yscale('log')
plt.xlabel('Iteraciones')
plt.ylabel('Error realtivo')
plt.title('Error realtivo de aproximar $ln(2)$  en $n$ iteraciones')
plt.show()
```
