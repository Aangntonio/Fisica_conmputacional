sea la función $f(x)$
$$
f(x) = \dfrac{1}{1-\tan hx}
$$


```python
f = lambda x: 1/(1- np.tanh(x))

plt.plot(x,f(x),color = '#FF4900')
plt.grid(True)
plt.title(' Gráfica de $ \dfrac{1}{1-tanhx}$')
plt.xticks([-2*np.pi,-1*np.pi,0,1*np.pi,2*np.pi],['$-2\pi$','$-\pi$','0','$\pi$','$2\pi$'])
plt.xlabel('x')
plt.ylabel('f(x)')
plt.show()
```

La función `tanh` o tangente hiperobólica definida puede interpretarse mediante la sigueinte gráfica:


```python
g = lambda x:  np.tanh(x)

x = np.linspace(-2*np.pi,2*np.pi,10001)
plt.plot(x,g(x))
plt.grid(True)
plt.title('$ tan h x$')
plt.yticks([-1,0,1])
plt.xticks([-2*np.pi,-1*np.pi,0,1*np.pi,2*np.pi],['$2\pi$','$\pi$','0','$\pi$','$2\pi$'])
plt.show()
```

Por lo que podemos ver en la función que para cuando $x \to -\infty$ entonces:
$$
\lim_{x \to -\infty} tanh(x) = -1
$$
Entonces:
$$
\lim_{x \to -\infty} \dfrac{1}{1-\tan hx} = \dfrac{1}{2}
$$
Sin embargo al para $x \to \infty$:
$$
\lim_{x \to \infty} \dfrac{1}{1-1} = \dfrac{1}{0}
$$
Lo cual supone un problema por la división entre 0
