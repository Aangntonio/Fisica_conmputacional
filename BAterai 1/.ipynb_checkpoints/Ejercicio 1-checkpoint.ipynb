La fórmula de la Ley de la Gravitación Universal de Isaac Newton en formato LaTeX:
$$
F_g = \frac{{G M  m_2}}{{r^2}}
$$
Donde:
Sin embargo por la segunda ley de newton se considera que :
$$
ma = \frac{{G Mm}}{{r^2}}
$$
Pero si consideramos una aceleracion causada por un movieminto circular .
$$
m \dfrac{v^2}{r} = \frac{{G M m}}{{r^2}}
$$
De esto se deduce que para una cierta r y dado que es un trayectoria eliptica :
$$

La ecuación polar que describe una órbita elíptica con el foco en el origen (el Sol) es:
$$
r = \frac{a(1 - e^2)}{1 + e \cdot \cos(\theta)}
$$
Donde:
- $a)$ es el semieje mayor de la elipse (la mitad de la distancia mayor entre el perihelio y el afelio).
- $e$ es la excentricidad de la elipse, que mide cuánto se desvía la órbita de ser una circunferencia perfecta.
- $\theta$ es el ángulo entre la posición actual del cometa y el perihelio.

Pero la excentricidad de la elipse se calcula como:

$$
e = 1 - \frac{r_{\text{afelio}}}{a}
$$
Pero dadas las condiciones del problema podemios ver que para el afelio la velocidad esta dad por:
$$
v_{\text{afelio}} = \sqrt{\frac{GM}{a}}
$$

$$
a = \frac{GM}{v_{\text{afelio}}^2}
$$
Sustituyendo \(a\) y \(e\) en la ecuación polar, obtendremos la ecuación que describe la posición del cometa en su órbita elíptica en función de \(\theta\):
$$
r = \frac{GM(1 - e^2)}{\left( v_{afelio}^2 \right)1 + e \cdot \cos(\theta)}
$$
QUe asu vez $\theta $ es una función del tiempo y que el periodo esta dado por 
$$

```python
import numpy as np

def distancia_orbital(theta):
    G = 6.67e11
    M = 1.99e30
    v = 9.12e2
    a = (G*M) / v**2
    r_a = 5.8e12
    e = 1 - (r_a/a)

    r = (G * M * (1 - e**2)) / ((v**2) * (1 + e * np.cos(theta)))
   
    return r
```

Por lo que se puede imeplentart una funcion ``conversion UA`` rapida que conveirta estos valores a `UA`


```python
def conversion_UA(distancia):
    '''
    Funcion que converierte distnacias medidas en m a UA
    '''
    return distancia / 149597870700 
```
