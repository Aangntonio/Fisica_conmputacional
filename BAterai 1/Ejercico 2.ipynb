```python
import math 
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Datos inicilaes
c= 1
v_i = 67
A = 0.67
rho = 1.27
g = 9.81

posiciones_x = []
posiciones_y = []

def tiro_vertical(t,angulo):
    theta = math.radians(angulo)
    v_x = v_i * np.cos(theta)
    v_y = v_i * np.sin(theta)
    x = v_x * t
    y = v_y * t - (g *  t**2)/2
    posicion = [x,y]  
    return posicion
```

```python
t_i = 0
t_f = 8

while t_i < t_f:
    pos = tiro_vertical(t_i,30)
    plt.scatter(pos[0],pos[1])
    t_i += 1
```
