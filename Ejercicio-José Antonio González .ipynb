```python
import numpy as np
import matplotlib.pyplot as plt 
#Se importan las bibliotecas
```

Para ver como se comporta el area bajo la curva de la sucesion:
$$
y_{n+1} = \int_0^{1} x^n e^x dx
$$

Donde se pudo deducir que :
$$
y_{n+1} = exp - (n+1) y_{n}
$$

entonces si consideramos si $n \to \infty $ entonces el area bajo la cruva tiende a 0, para verlo usamos las paqueterias de `matplotlib.pyplot` y el modulo `numpy`

```python
def succesion(n):
    '''
    La función genera la sucesion de la expresion x^n exp(x) tomando como argumento el parametro n
    '''
    
    x = np.linspace(0,1,101) # se define el intervalo de integración
        
    f = lambda x: x** n * np.exp( x)
    plt.plot(x,f(x),label = 'n = ' + str(n)) # Con el modulo de plt se hace la grafica tomando como leyenda el valor de cada iteración
    plt.legend()
    plt.xlim(0,1) #Dado que es un intervalo cerrado se define la función en solo esa región
    plt.xlabel('$x$')
    plt.ylabel('$f(x)$')
```

```python
# Una vez obtenida la función que genera las graficas, iteramos con un ciclo 
for i in range (1,10,1):
    if i == 1 or i%2 == 0 :
        succesion(i)
```
