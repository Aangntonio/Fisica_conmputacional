```python
def barkara(a,b,c):
    '''
    Esta funci√≥n puede resolver una ecuacion cuadratica con :
    a: 
    b:
    c:
    '''
    if b**2- 4 * a * c <0:
        raise ValueError('La solucion no esta definida en R')
    if a == 0:
        raise ValueError('a debe ser distindo de 0')
        
    dis = np.sqrt(b**2 - 4*a*c)
    x1 = (-b+  dis)/ (2*a)
    x2 = (-b-  dis)/(2*a)
    
    f = lambda x: a*x**2 + b*x + c 
    x = np.linspace(-10,10,101)
    
    plt.plot(x,f(x))
    plt.axhline(0, color='black', linewidth=0.5)
    plt.axvline(0, color='black', linewidth=0.5)
    plt.scatter(x1, 0, color='red', marker='o')
    plt.scatter(x2, 0, color='red', marker='o')
    plt.grid()
    plt.show()
    
    
    return 'las soluciones a tu ecuacion son', x1,' y', x2
```

```python
for i in range(1,10,2):
    barkara(-i,1/i,i)
    #barkara(i,0,-i)
```
