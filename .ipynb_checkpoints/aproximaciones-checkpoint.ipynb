```python
import numpy as np 
from math import factorial as factorial
import matplotlib.pyplot as plt
```

```python
def sin(x,n = 64):
    '''
    Esta función calcula el seno con un n veces de interaciones en la serie de Taylor
    '''
    if type(n)!= int :
        raise ValueError("Debes introducir un numero real")
    if n < 0:
        raise ValueError("Debe introcur un numero mayor que 0")
    x_real = float(x)
    i = 0
    suma = 0
    while i < n:
       
        suma  = suma + ((-1)**i)*(x_real**(2*i-1))/ (factorial(2*i+1))
        i = i +1
    return suma
```

```python
import math

math.sin(np.pi/6 )
        
```

```python
def error(arg1,arg2):
    '''
    '''
    return  abs((arg1-arg2)/arg1) * 100
a = float(input('introduce el número: '))
x = math.radians(a)
j = 0
n = 10
suma = x
termino = x
lista_j = []
lista_suma = []
lista_error = []

print('iteraciones \t approx \t error')
print('-'*40)

for i in range(2,n):
    j+=1
    termino = (-termino*x*x)/( (2* i -1)* (2 * i -2)  ) 
    suma = suma + termino
    lista_j.append(j)
    lista_suma.append(suma)
    lista_error.append(error(math.sin(x),suma))
    
    print('{0:} \t \t {1:1.10f} \t {2:1.5e}'.format(j,suma,error(math.sin(x),suma)))
    

plt.plot(lista_j,lista_error,'r+')
plt.plot(lista_j,lista_error,'--', color = '#0C687F')
plt.xlabel('iteracion')
plt.ylabel('Error')
cadena = 'Error relativo al calcular $sin (x)$'
plt.title(cadena)
plt.show()

plt.plot(lista_j,lista_suma,'bo--')
plt.xlabel('iteracion')
plt.ylabel('Valor')
cadena2 = 'aproxiamcion de $sen ($' + str(a) + '$)$'
plt.title(cadena2)
plt.show()

print(lista_error)
```
