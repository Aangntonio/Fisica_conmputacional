```python
import numpy as np
import matplotlib.pyplot as plt
from numpy import sqrt as sqrt
```

```python
t = []
v = []

dt = 1

Potencia = 400
masa = 70
tmax =200
nmax = tmax / dt

t.append(0)
v.append(4)

for i in range(int(nmax)):
    ti = t[i-1] + dt
    vi = sqrt(v[i]**2 + (2*Potencia* dt )/masa )
    
    t.append(ti)
    v.append(vi)

plt.plot(v,'-',color = 'red')
plt.xlabel('tiempo[s]')
plt.xlim([0,200])
plt.ylabel('velocidad $[\dfrac{m}{s} ]$ ')
plt.title('Velocidad de un ciclista sin resistencia al aire')
plt.show()
lista_velocidad = v
```
